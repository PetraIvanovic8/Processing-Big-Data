```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('monthlyResults.csv')

for col in df.columns:
    try:
        df[col] = df[col].astype(float)
    except ValueError:
        try:
            df[col] = df[col].astype(int)
        except ValueError:
            # If conversion to both float and int fails, the column will remain unchanged
            print(f"Column '{col}' cannot be converted to int/float.")

df.info()
```

```python
correlation_matrix = df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.show()
```
